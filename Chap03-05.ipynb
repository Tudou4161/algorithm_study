```
#90페이지. 최소 거스름 돈 동전 개수 출력하기

def remain_coin(remain, coin_list):
    cnt = 0
    for coin in coin_list:
        cnt += remain // coin #거스름돈에서 동전 종류 중 하나를 나눈 값을 cnt에 저장한다.
        print(cnt)
        remain %= coin #거스름 돈에서 동전 종류 중 하나를 나눈 값의 나머지 값을 remain에 재할당한다.
        print(remain)

    return f"최소 거스름 돈 동전 개수: {cnt}"


if __name__ == "__main__":
    remain = int(input("거스름 돈 입력:"))
    coin_list = [500, 100, 50, 10]
    print(remain_coin(remain, coin_list))
```

```
# 92p. 큰 수의 법칙
def Big_integer_rule(N, M, K, integer_list):
    integer_list.sort()
    first = integer_list[-1] #정렬 후 제일 큰 수를 저장
    second = integer_list[-2] #정렬 후 두 번째로 큰 수를 저장
 
    count = int( M / (K+1) ) * K  #큰 수가 연산될 횟수를 미리 파악한다.
    count += int(M % (K+1)) #홀수일 것을 대비해, 나머지 값을 cnt에 재차 할당한다.

    res = 0    #최종 리턴 값 초기엔 0으로 선언
    res += count * first  #위에서 정의한 cnt값에 첫 번째로 큰 수를 곱해준다.
    res += (M - count) * second  #남은 cnt값을 바탕으로, 두 번째로 큰 수를 곱해준다.
    return res

if __name__ == "__main__":
    N, M, K = map(int, input().split())
    integer_list = list(map(int, input().split()))
    print(Big_integer_rule(N, M, K, integer_list))
```

```
#96p.숫자카드게임
#각 행 별 최소 숫자를 구하고, 그 최소숫자의 조합 중 가장 큰 정수를 출력하면 된다.

def card_game(rows, cols):
    result = 0
    for _ in range(rows):
        min_int = min(list(map(int, input().split())))  #각 행 별 최솟값을 min_int변수에 담아준다.
        in_max = max(result, min_int) #굳이 리스트를 할당받지말고, max함수를 사용해서 해결해보자. 위에서 선언한 result변수를 적극 활용해본다.
        result = in_max 
    return result

if __name__ == "__main__":
    rows, cols = input().split()
    print(card_game(int(rows), int(cols)))
```

```
#100페이지. 1이 될 때 까지
def to_one(n, k):
    res = 0
    while True:
        target = (n // k) * k #k로 나누어 떨어질 정수를 미리 파악하기 위해서, 다음과 같이 연산한다. #패턴파악이 매우 중요하다.
        print("target: ", target)
        res += (n - target) #res에 n - target(1로 빼야하는 횟수)를 미리 저장한다.
        print("res: ", res)
        n = target #타깃값을 n에 할당한다.

        if n < k: #타깃값보다 k가 커지는 순간
            break; #알아서 멈춘다.
        res += 1
        n //= k
    
    res += (n - 1) #타깃을 의도적으로 0으로 만들고, -1해주면 올바른 결과가 출력된다.
    return res

if __name__ == "__main__":
    n, k = map(int, input().split())
    print(to_one(n, k))
```

```
#110p 상하좌우 문제
# 2차원 공간상황을 가정하는 문제들이 많이 출제된다.
def updown(N, plans):
  x, y = 1, 1 #초기값
  dx = [0, 0, -1, 1] #무빙 타입에 해당하는 움직임 값
  dy = [-1, 1, 0, 0] #x와 비슷
  move_types = ["L", "R", "U", "D"] #움직임 타입 #좌, 우, 상, 하

  for plan in plans:
    for i in range(len(move_types)):
      if plan == move_types[i]:
        nx = x + dx[i]
        ny = y + dy[i]

    if nx < 1 or ny < 1 or nx > N or ny > N:
      continue   #만약 if문에 걸리는 값이 되면, 아래의 x, y에 신규 값을 재할당하지 않는다. 고로 x,y는 둘 중 하나는 아직 1에 머물러있는 상태

    x, y = nx, ny
  return x, y

if __name__ == "__main__":
  N = int(input())
  plans = input().split()
  x, y = updown(N, plans)
  print(x, y)
```

```
#113p. 시각 문제 -> 모든 시, 분, 초에 N이 포함되는 모든 경우의 수를 출력한다.
def time_cnt(N):  # -> N이 3이면, 0시00분00초부터 3시59분59초까지 3이 출력될 모든 경우의 수를 출력하면 된다.
    cnt = 0
    for i in range(N+1):
        for j in range(60):
            for k in range(60):
                if str(N) in str(i) + str(j) + str(k):
                    cnt += 1
    return cnt

if __name__ == "__main__":
    N = int(input())
    cnt = time_cnt(N)
    print(cnt)
```

```
#115p 왕실의 나이트 문제

def knightsOfkingdom(input_data):

    row = int(input_data[1]) # 입력 문자열의 2번째 원소 
    col = int(ord(input_data[0])) - int(ord("a")) + 1 #입력 문자열의 1번째 원소
    
    steps = [  
             (-2, 1), (-1, -2), (1, -2), (2, -1),
             (2, 1), (1, 2), (-1, 2), (-2, 1)
    ]  #나이트가 갈 수 있는 모든 경우의 수

    result = 0
    for step in steps: 
        next_row = row + step[0] #현재 행에서 스텝 추가
        next_col = col + step[1] #현재 열에서 스텝 추가
        if next_row >= 1 and next_col <= 8 and next_row <= 8 and next_col >= 1:
            result += 1 
    return result

if __name__ == "__main__":
    input_data = input()
    print(knightsOfkingdom(input_data))
```

```
##문자열 재정렬 문제 -> 앞에서부턴 알파벳을 정렬해주고, 뒤에선 숫자의 합계를 출력한다.

def string_sort(input_str):
    alpha_list = [] #알파벳 문자열의 총합을 구한다.
    total_int = 0 #숫자 문자열의 총합을 구한다.
    for i in range(len(input_str)):
        if input_str[i].isalpha() == True: #만약 알파벳이라면
            alpha_list.append(input_str[i]) #별도의 리스트에 append
        else:
            total_int += int(input_str[i]) #숫자 문자열 합계 총합
    
    alpha_list.sort()
    alpha_join = ''.join(alpha_list)
    return alpha_join + str(total_int) #정렬된 알파벳과 숫자 합계를 merge한다.

if __name__ == "__main__":
    input_str = input()
    print(string_sort(input_str))
```

```
#간단한 재귀함수 구현
def recursive_func(i):
    #기저조건 설정 매우 중요!
    if i == 10:
        return
    print(f"{i}번째 재귀함수에서 {i+1}번째 재귀함수를 호출합니다.")
    recursive_func(i + 1) #총 i번만큼 호출을 진행한다.

    print(f"{i}번째 재귀함수를 종료합니다.") #호출된 함수를 하나씩 불러 출력한다.

if __name__ == "__main__":
    i = int(input())
    print(recursive_func(i))
```

```
#팩토리얼을 재귀함수로 구현
def factorial(i):
    if i <= 1:
        return 1
    return i * factorial(i-1)

if __name__ == "__main__":
    print(factorial(5))
```

```
#유클리드 호제법 계산
def gcd(a, b):
    if a % b == 0: #a를 b로 나눈 값이 0이면 b를 리턴한다.
        return b
    else:
        return gcd(b, a % b) #재귀호출한다!


print(gcd(192, 162))
```

```
#dfs 예제 코드
def dfs(graph, v, visited):
    visited[v] = True  #첫번째 방문지는 일단 True로 바꿔서 방문처리한다. #트루 -> 방문상태 / 펄스 -> 방문하지 않은상태/
    print(v, end=' ')  
    for i in graph[v]:
        if not visited[i]:
            dfs(graph, i, visited)

graph = [
         [],
         [2,3,8],
         [1,7],
         [1,4,5],
         [3,5],
         [3,4],
         [7],
         [2,6,8],
         [1,7]
]

visited = [False] * 9
dfs(graph, 1, visited)
```

```
#BFS예제 코드
from collections import deque

def bfs(graph, start, visited):
    queue = deque([start]) #1을 큐에 담아둔다.
    visited[start] = True #1을 방문처리한다.
    while queue: #queue가 빌 때까지 반복한다. 계~~~속 반복한다.
        v = queue.popleft() #왼쪽부터 원소를 꺼낸다 (1 popout)
        print(v, end=' ') 
        for i in graph[v]: #그래프의 1번 내부리스트의 원소를 하나씩 출력한다.
            if not visited[i]: #만약 방문처리 되지 않았다면..
                queue.append(i) #큐에 삽입한다.
                visited[i] = True #그리고 방문처리한다. 다음 과정을 계속 반복한다.


graph = [
         [],
         [2,3,8],
         [1,7],
         [1,4,5],
         [3,5],
         [3,4],
         [7],
         [2,6,8],
         [1,7]
]


visited = [False] * 9

bfs(graph, 1, visited)
```

```
#음료수 얼려먹기 -> 한 번에 만들 수 있는 아이스크림 갯수 출력하기
n, m = map(int, input().split())
graph = []
for i in range(n):
    graph.append(list(map(int, input())))

def dfs(x, y):
    if x <= -1 or x >= n or y <= -1 or y >= m: #기저조건 처리
        return False

    if graph[x][y] == 0:
        graph[x][y] = 1
        dfs(x-1, y)
        dfs(x, y-1)
        dfs(x+1, y)          #아...진짜 이해할 수가 없네.....후 ,,,^^
        dfs(x, y+1)
        return True
    return False

result = 0
for i in range(n):
    for j in range(m):
        if dfs(i, j) == True:
            result += 1

print(result)
```

```
#미로탈출
#공백 구분해서 입력받기 / 그래프도 루프 돌려서 입력받기
#이동방향 정의
#BFS함수 구현
#첫 번째 원소 꺼내서 큐가 빌때까지 계속 반복한다. while루프
#현재 위치에서 4방향 확인 -> 공간 벗어나거나 벽인 경우 해당 노드를 처음 방문하는 경우에만 최단 거리 기록
#가장 오른족 아래까지의 최단 거리 기록

from collections import deque

n, m = map(int, input().split())
graph = []
for _ in range(n):
    graph.append(list(map(int, input())))
dx = [-1, 1, 0, 0]
dy = [0, 0, -1, 1]


def BFS(x, y):
    queue = deque()  #덱 importt시켜준다.
    queue.append((x,y)) #튜플형태로 값을 받아온다. 초기 위치를 저장한다. 보통 0,0 일 것이다.

    while queue: #큐가 빌때까지 계속 루프를 돌린다.
        x, y = queue.popleft() #맨 왼쪽 원소를 pop해준다.
        for i in range(4):   #4개 방향 모두 탐색하기 위해서 루프를 돌려준다..
            nx = x + dx[i]   #위치 증감 ㄱㄱㄱ
            ny = y + dy[i]   #위치 증감 ㄱㄱㄱ

            if nx < 0 or ny < 0 or nx >= n or ny >= m: #xy가 0보다 작거나, 그래프 밖을 벗어나게 된다면 ....
                continue #쌩까고 넘어간다.
            if graph[nx][ny] == 0: 
                continue #벽을(0)을 마주치면, 마찬가지로 쌩깐다 ^^
            if graph[nx][ny] == 1: #벽이 아니라면...
                graph[nx][ny] = graph[x][y] + 1 #벽이 아닌 위치에 현재까지 이동한 횟수를 저장하기 위해서 증감연산을 실시한다 !
                queue.append((nx, ny)) #그리고, 해당 원소를 다시 큐에 삽입한다.
    return graph[n-1][m-1]   #우측 하단, 하측 우단에 찍힐 최종 이동횟수 값을 리턴시켜준다!

print(BFS(0,0))
```
